```python
#MEMBUAT MODEL MENGGUNAKAN tf.data.Datasets dari MNIST

import tensorflow as tf
mnist = tf.keras.datasets.mnist

(image_train, label_train), (image_test, label_test) = mnist.load_data()
```

```python
#menampilkan label dr salah 1 gbr data train dg matplotlib

import numpy as np
np.set_printoptions(linewidth=200)
import matplotlib.pyplot as plt
plt.imshow(image_train[0])
print(label_train[0])
print(image_train[0])
```

```python
#normalisasi dataset
image_train = image_train/255.0
image_test = image_test/255.0

'''
dlm proses pengolahan citra atau pengenalan pola, normalisasi gbr
penting dilakukan krn utk memastikan bhw nilai piksel berada dlm
rentang yg dpt diolah dg baik oleh model. dlm hal ini, kode tsb
membagi nilai piksel dr gbr latih (image_train) dg 255.0

nilai piksel pd umumnya berada dlm rentang 0-255, pembagian dg nilai
255.0 akn menghasilkan nilai piksel yg lbh kecil, dr rentang  0-1. hal ini
dpt membantu model machine learning utk belajar lbh efisien krn nilai
yg lbh kecil cenderung memiliki efek yg lbh stabil selama proses learning.'''

```

```python
#MEMBANGUN ARSITEKTUR DG 3 LAYER

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28)),
    tf.keras.layers.Dense(128, activation = tf.nn.relu),
    tf.keras.layers.Dense(10, activation = tf.nn.softmax)
])
```

```python
#menentukan optimizer dan loss utk model dan lakukan training

model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
model.fit(image_train, label_train, epochs = 5)
```

```python
#PRACTICE DG DATASET LAIN, cifar100

import tensorflow as tf
cifar100 = tf.keras.datasets.cifar100

(image_train, label_train), (image_test, label_test) = cifar100.load_data()
```

```python
#menampilkan label dari salah 1 data cifar100

import numpy as np
np.set_printoptions(linewidth = 200)
import matplotlib.pyplot as plt
plt.imshow(image_train[0])
print(label_train[0])
print(image_train[0])
```

```python
#normalisasi dataset
image_train = image_train/255.0
image_test = image_test/255.0
```

```python
#membuat arsitektur dg 3 layer

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation = 'relu', input_shape=(32,32,3)),
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Conv2D(64, (3,3), activation ='relu'),
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(100, activation = 'softmax')
])
```

```python
#tambahkan dropout
tf.keras.layers.Dropout(0.25)
```

```python
#data augmentation

datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range = 30,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    horizontal_flip = True
)
datagen.fit(image_train)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
```

```python
tfds.list_builders()
```

```python
ds  = tfds.load('mnist', split = 'train', shuffle_files = True)
assert isinstance(ds, tf.data.Dataset)

'''
penjelasan dari parameter di atas:
- fungsi load(): memungkinkan utk menggunakan dataset yg telah provide TensorFlow Datasets
- parameter('mnist'):nama dataset yg ingin dimuat
- parameter('split = train'): bhw kita memuat bag. pelatihan dr dataset MNIST
- parameter('shuffle_files'=true):bhw file2 dataset akn diacak sblm dimuat,
shg. urutan data tdk akn ttp sm setiap kli dataset dimuat ulang.
'''
```

```python
#LATIH DATA
(train_images, train_labels), (test_images, test_labels) = tfds.as_numpy(tfds.load('mnist',
                              split = ['train', 'test'],
                              batch_size = -1,
                              as_supervised = True))
```

```python
#MEMBANGUN ARSITEKTUR MODEL

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation = 'relu', input_shape = (28, 28, 1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64,(3,3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model.compile(
    optimizer = tf.keras.optimizers.RMSprop(),
    loss = 'sparse_categorical_crossentropy',
    metrics=['accuracy'],
)
```

```python
model.fit(train_images, train_labels, batch_size = 50, epochs=5)
```

```python
model.evaluate(test_images, test_labels, batch_size = 1)
```
